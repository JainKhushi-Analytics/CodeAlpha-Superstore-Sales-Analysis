## Sales by category

```python

sales_by_category=df.groupby('Category')['Sales'].sum()
```

```python
sales_by_category
```

```python
plt.figure()
plt.bar(sales_by_category.index,sales_by_category.values,color='blue',alpha=0.5)
plt.title('Sales By Category')
plt.xlabel('Category')
plt.ylabel('Tatal Sales')
plt.tight_layout()
plt.show()
```

## profit by region

```python
profit_by_region=df.groupby('Region')['Profit'].sum()
```

```python
profit_by_region
```

```python
#profit by region
plt.figure()
plt.bar(profit_by_region.index,profit_by_region.values,color='blue',alpha=0.5)
plt.title('Profit By Profit')
plt.xlabel('Region')
plt.ylabel('Tatal Sales')
plt.tight_layout()
plt.show()
```

## #Top 10 sub category by sales


```python
top_10_subcategory=df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
top_10_subcategory
```

```python
plt.figure()
plt.barh(top_10_subcategory.index,top_10_subcategory.values,color='blue',alpha=0.5)
plt.title('Top 10 Sub_Category')
plt.xlabel('Sales')
plt.ylabel('Sub Category')
plt.tight_layout()
plt.show()
```

## Loss making sub category

```python

loss_subcategory=df.groupby('Sub-Category')['Profit'].sum().sort_values().head(5)
loss_subcategory
```

```python
plt.barh(loss_subcategory.index,loss_subcategory.values,color='blue',alpha=0.5)
plt.title('Top 5 Loss Making Sub_Category')
plt.xlabel('Loss')
plt.ylabel('Sub Category')
plt.tight_layout()
plt.show()
```

## sales vs profit

```python
sales_profit=df.groupby('Sales')['Profit'].mean()
sales_profit
```

```python
plt.figure(figsize=(7,8))
plt.scatter(sales_profit.index,sales_profit.values,marker='o',color='blue',alpha=0.3)
plt.axhline(0)
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('profit')
plt.show()
```
